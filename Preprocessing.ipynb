```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

`pandas` buat ngolah data tabular (CSV).

`numpy` bantu operasi numerik.

`LabelEncoder dan MinMaxScaler` dari sklearn dipakai nanti buat encoding dan normalisasi data.

```python
df = pd.read_csv("bps-od_15042_indeks_pmbngnn_manusia__prov_di_indonesia_data.csv")
df.head()
```

Baris ini buat membaca file CSV ke bentuk DataFrame (df).

`df.head()` nunjukin 5 baris pertama biar kita tahu isi dan struktur data.

```python
df.info()
```

Perintah ini buat lihat jumlah baris, kolom, dan tipe data tiap kolom `(numerik, objek, float, int, dll).`

Berguna buat tahu kolom mana yang mungkin harus diubah tipenya nanti.

```python
df.describe()
```

Ngasih ringkasan statistik dasar `(mean, min, max, dll)` untuk semua kolom numerik.

Bisa bantu kamu lihat ada nilai yang aneh atau outlier.

```python
print("Jumlah missing value tiap kolom:")
print(df.isnull().sum())

# Hapus baris yang kosong
df.dropna(inplace=True)
```

`df.isnull().sum()` buat lihat apakah ada data kosong di tiap kolom.

`Kalau ada, dropna()` dipakai buat hapus baris kosong supaya gak ganggu proses analisis selanjutnya.

```python
print("Jumlah data duplikat:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

Bagian ini buat ngecek apakah ada data yang dobel `(baris sama persis)`

dan langsung hapus pakai `drop_duplicates()` biar datanya bersih.

```python
df['tahun'] = df['tahun'].astype(int)
```

Kolom “Tahun” sering kali terbaca sebagai float `(misal 2010.0)`,
jadi diubah jadi int biar lebih rapi dan sesuai makna datanya.

```python
le = LabelEncoder()
df['nama_provinsi'] = le.fit_transform(df['nama_provinsi'])
```

Kolom “Provinsi” berisi teks, sedangkan model machine learning cuma bisa baca angka.
Makanya, kita ubah jadi bentuk numerik dengan Label Encoding, misalnya:

Jawa Tengah → 0

Jawa Barat → 1

dst.

```python
scaler = MinMaxScaler()

kolom_numerik = ['indeks_pembangunan_manusia']

df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])
```

Langkah ini buat menyamakan skala data numerik ke rentang 0–1.

Karena beberapa algoritma (misal KNN, Decision Tree, dll) sensitif sama perbedaan skala data.

`MinMaxScaler` bantu ngelakuin secara otomatis.

Buat memastikan hasil preprocessing udah jalan — kolom “Provinsi” udah jadi angka.

```python
df.to_csv("ipm_preprocessed.csv", index=False)
print("Data berhasil disimpan sebagai ipm_preprocessed.csv")
```
